Build a regression model.

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('df_merged_clean_csv.csv')

df_model = df[['station_id', 'station_name', 'total_number_bikes']].drop_duplicates() # I keep one row per station here 

number_of_pois = df.groupby('station_id')['poi_name'].count()
df_model['number_of_pois'] = df_model['station_id'].map(number_of_pois) 

avg_rating = df.groupby('station_id')['rating'].mean()
df_model['avg_rating'] = df_model['station_id'].map(avg_rating)

df_model.head()

df_model.isnull().sum() # No nulls left in the model
df_model.duplicated().sum() # No duplicates left in the model

sns.histplot(df_model['number_of_pois'], bins = 10, kde = False) 
plt.show()
sns.histplot(df_model['avg_rating'], bins = 10, kde = False)  
plt.show()
sns.boxplot(data = df_model[['total_number_bikes', 'number_of_pois', 'avg_rating']])
plt.show()
# Here we see some outliers left for total_number_bikes and number_of_pois that we could fix before moving on to the model fitting.
# Since avg_rating is capped and the all values are between 1 and 5, I will not transform based on that part of the boxplot.
df_model.sort_values(by='total_number_bikes', inplace = True)
Q1 = df_model['total_number_bikes'].quantile(0.25)
Q3 = df_model['total_number_bikes'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_model = df_model[(df_model['total_number_bikes'] >= lower_bound) & (df_model['total_number_bikes']<= upper_bound)]

df_model.sort_values(by='number_of_pois', inplace = True)
Q1 = df_model['number_of_pois'].quantile(0.25)
Q3 = df_model['number_of_pois'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_model = df_model[(df_model['number_of_pois'] >= lower_bound) & (df_model['number_of_pois']<= upper_bound)]
```

```python
df_model.sort_values(by = 'station_id', inplace = True) # re-ordering the Dataframe and seeing how the data left looks like
df_model.head()
df_model.shape

sns.heatmap(df_model[['total_number_bikes', 'number_of_pois', 'avg_rating']].corr(), annot = True, cmap = 'viridis')
plt.show()
# This heatmap indicates a very weak linear relationship between our variables, which likely implies a weak future model.
```

```python
X = df_model[['number_of_pois', 'avg_rating']]
X = sm.add_constant(X)
y = df_model['total_number_bikes']

multi_reg = sm.OLS(y, X)
reg_model = multi_reg.fit()
```

Provide model output and an interpretation of the results. 

```python
print_model = reg_model.summary()
print(print_model)
# This summary shows an Adj. R-Squared value of -0.003 and p-values of the variables of 0.716 and 0.239, so we can say with certainty that this model is a poor fit
# and that the variables do not have a strong relationship with the total number of bikes here with values largely above 0,05. 
# We could remove the largest p-value variable here being number_of_pois but this action would not drastically improve the model given the very low overall Adj.R Squared.
# We could also try other models or consider data scaling or transformations given the distribution of the two independent variables, but even that might not improve significantly our model.
```
