```python
import os
import requests
import pandas as pd
from pprint import pprint
```

# Foursquare

```python
Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 
```

```python
FOURSQUARE_API_KEY = os.getenv('FOURSQUARE_API_KEY')
```

```python
FOURSQUARE_API_KEY = os.getenv('FOURSQUARE_API_KEY')

df_bikes_paris = pd.read_csv('bikes_paris_csv.csv')

latitude = df_bikes_paris.loc[0,'latitude']
longitude = df_bikes_paris.loc[0, 'longitude']

headers = {
    'Accept': 'application/json',
    'Authorization' : FOURSQUARE_API_KEY
}
url = 'https://api.foursquare.com/v3/places/search'
params = {
    'll' : f'{latitude},{longitude}',
    'radius' : 1000, 
    'categories' : '4d4b7105d754a06374d81259', # restaurants
    'limit' : 10
}

response = requests.get(url = url, headers = headers, params = params)

if response.status_code == 200:
    data = response.json()
else:
    print(f'{response.status_code}')
```

```python
data['results'][0]
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
stations_data_foursquare = []

for i in range(len(df_bikes_paris)):
    latitude = df_bikes_paris.loc[i,'latitude']
    longitude = df_bikes_paris.loc[i, 'longitude']
    params = {
    'll' : f'{latitude},{longitude}',
    'radius' : 1000, 
    'categories' : '4d4b7105d754a06374d81259,4bf58dd8d48988d16d941735,4bf58dd8d48988d181941735', # restaurants, cafes and museums
    'limit' : 50 }
    response = requests.get(url = url, headers = headers, params = params)
    if response.status_code == 200:
        data = response.json()
    
        for place in data.get('results', []):
            place_info = {
                'station' : i,
                'name' : place.get('name'),
                'distance' : place.get('distance'),
                'address' : place.get('location',{}).get('address'),
                'category' : place.get('categories',[{}])[0].get('name')
            }
            stations_data_foursquare.append(place_info)
    else:
        print(f'{response.status_code}')
```

Put your parsed results into a DataFrame

```python
df_foursquare = pd.DataFrame(stations_data_foursquare)
df_foursquare.to_csv('df_foursquare_csv.csv', index = False)
df_foursquare.head()
```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
YELP_API_KEY = os.getenv('YELP_API_KEY')

latitude = df_bikes_paris.loc[0,'latitude']
longitude = df_bikes_paris.loc[0, 'longitude']

headers_yelp = {
    'Accept': 'application/json',
    'Authorization' : f'Bearer {YELP_API_KEY}'
}
url_yelp = 'https://api.yelp.com/v3/businesses/search'
params_yelp = {
    'latitude' : latitude,
    'longitude' : longitude,
    'radius' : 1000, 
    'categories' : 'italian,japanese,thai,french,mexican,cafes,museums',
    'limit' : 10
}

response_yelp = requests.get(url = url_yelp, headers = headers_yelp, params = params_yelp)

if response_yelp.status_code == 200:
    data_yelp = response_yelp.json()
else:
    print(f'{response_yelp.status_code}')

data_yelp['businesses'][0]
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
stations_data_yelp = []

for i in range(len(df_bikes_paris)):
    latitude = df_bikes_paris.loc[i,'latitude']
    longitude = df_bikes_paris.loc[i, 'longitude']
    params_yelp = {
    'latitude' : latitude,
    'longitude' : longitude,
    'radius' : 1000, 
    'categories' : 'italian,japanese,thai,french,mexican,cafes,museums',
    'limit' : 50 }
    response_yelp = requests.get(url = url_yelp, headers = headers_yelp, params = params_yelp)
    if response_yelp.status_code == 200:
        data_yelp = response_yelp.json()
    
        for place in data_yelp.get('businesses', []):
            place_info = {
                'station' : i,
                'name' : place.get('name'),
                'distance' : place.get('distance'),
                'address' : place.get('location',{}).get('address1'),
                'rating' : place.get('rating')
            }
            stations_data_yelp.append(place_info)
    else:
        print(f'{response_yelp.status_code}')
```

Put your parsed results into a DataFrame

```python
df_yelp = pd.DataFrame(stations_data_yelp)
df_yelp.to_csv('df_yelp_csv.csv', index = False)
df_yelp.head()
```

# Comparing Results

Which API provided you with more complete data? Provide an explanation. 

```python
print(df_foursquare.shape)
print(df_yelp.shape)
# more POI in the Yelp Database
print(df_foursquare.groupby('station').size().mean())
print(df_yelp.groupby('station').size().mean())
# Here again there is in average more POI per station in the Yelp Database
print(df_foursquare.describe())
print(df_yelp.describe())
# It seems that in general, the POIs are closer to the station in the Foursquare Dataframe, where in the Yelp Dataframe even with radius set to 100, a lot of the data is beyong that
print(df_foursquare.isnull().sum())
print(df_yelp.isnull().sum())
# The Foursquare DataFrame have a bit more missing values with 6 addresses missing, with only 2 in the df_yelp
# Yelp provides ratings of the different POIs, where Foursquare doesn't
# Even though Yelp doesn't provide the category so easily and clearly as Foursquare, Yelp still overall provides with more complexe data, so I will continue my analysis with the Yelp data
```
